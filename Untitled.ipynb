```python
import uuid #unique identifier
import os 
import time 
```

```python
IMAGES_PATH= os.path.join('data','images') #save image data, topfolder, subscriptfolder
labels=['awake','drowsy'] #classes
number_imgs= 20 #number of images that we wanna collect
```

```python
#collecting images basically creating a small dataset

cap=cv2.VideoCapture(0)
for label in labels: #looping through each label here drowsy and awake
    print('Collecting images for {}'.format(label))
    time.sleep(5)
    #loop through image range
    for img_num in range(number_imgs): #loop through each img
        print('Collecting images for {}, image number {}'.format(label, img_num)) 
        ret, frame= cap.read() #read the frame from our webcam
        
        imgname= os.path.join(IMAGES_PATH, label+'.'+str(uuid.uuid1())+'.jpg') #created path of image data
        cv2.imwrite(imgname, frame) #writes out img to file
        cv2.imshow('Image Collection', frame) #render to the screen
        time.sleep(2) #2sec delay betn captures
        
        #exiting
        if cv2.waitKey(10) & 0xFF == ord('q'):
            break
cap.release()
cv2.destroyAllWindows()
    
```
